```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.utils import resample
```

```python
df = pd.read_csv("diabetes.csv")
```

```python
pos = df.Insulin[df.Outcome == 1]
neg = df.Insulin[df.Outcome == 0]
len(pos), len(neg)

```

```python
plt.hist(neg, bins=10, color='red', alpha=0.5, label='Negative')
plt.hist(pos, bins=10, color='blue', alpha=0.4, label='Positive')

plt.legend()
plt.savefig("1.svg")
plt.show()
```

```python
pos = resample(
    pos,
    replace=True,
    n_samples=len(neg),
)

len(pos), len(neg)
```

```python
np.average(pos), np.average(neg)
```

```python
np.average(pos), np.average(neg)
```

```python
plt.hist(neg, bins=10, color='red', alpha=0.5, label='Negative')
plt.hist(pos, bins=10, color='blue', alpha=0.4, label='Positive')

plt.legend()

plt.savefig("2.svg")
plt.show()
```
