```python
import matplotlib.pyplot as plt
import numpy as np
from matplotlib import hatch
```

```python
x = ["jan", "feb", "mar", "apr", "may", "jun", "jul", "aug", "sep", "oct", "nov", "dec"]
data_it = [7, 19, 3, 42, 11, 28, 5, 16, 9, 31, 24, 14]
data_mtk = [18, 4, 27, 13, 35, 2, 21, 40, 8, 29, 15, 6]
```

```python
def getRange(*args):
    maxVal = np.max(args)
    minVal = np.min(args)
    return range(int(np.ceil(minVal)), int(np.ceil(maxVal*1.1)), 10)
```

```python
getRange(data_it, data_mtk)
```

```python
#                    h  v
plt.figure(figsize=(20, 8))

plt.plot(x, data_it, label="IT", color="#FF00FF")
plt.plot(x, data_mtk,
         label="Mtk",
         linestyle="-.",
         color="#F0F0F0",
         linewidth=2,
         marker="s",
         markerfacecolor="#FF0000",
         markeredgecolor="#FF0000",
         markersize=10
       )
plt.ylabel("Vendas")
plt.xlabel("mes")
plt.legend()
plt.title("Vendas do ano de 2025")
plt.grid(True, linestyle="--")


plt.yticks(getRange(data_it, data_mtk))
plt.savefig("Vendas.svg")
plt.savefig("Vendas.png")
plt.savefig("Vendas.jpg")
plt.show()
```

```python
plt.figure(figsize=(20, 8))

plt.bar(x,
        data_it,
        label="IT",
        color="#FF00FF",
        width=0.6,
        edgecolor="#FF0000",
        linewidth=2,
        hatch="*",
        )

plt.ylabel("Vendas")
plt.xlabel("mes")
plt.legend()
plt.title("Vendas do ano de 2025")
plt.show()
```

```python
plt.figure(figsize=(8, 20))

plt.barh(x,
        data_it,
        label="IT",
        color="#2abf9f",
        hatch="*",
        )

plt.title("Vendas do ano de 2025", fontweight="bold", fontsize="18")
plt.ylabel("Vendas", fontweight="bold", fontsize="18")
plt.xlabel("mes", fontweight="bold", fontsize="18")

plt.legend()
plt.show()
```

```python
horas_estudo = [4,4,3,1,4,6,4,4, 1]
notas = [12,14, 11,8,13,19,1,20,17]

x = np.array(horas_estudo)
y = np.array(notas)
coef = np.polyfit(x,y,1)
tendencia = np.poly1d(coef)


plt.figure(figsize=(10, 8))

plt.plot(x, tendencia(x), color="#FF0000")

plt.scatter(x,
            y,
            s=122,
            color="#FFFFFF",
            alpha=0.5,
            label="Masculino",
            marker="^",)

plt.legend()

plt.show()
```
