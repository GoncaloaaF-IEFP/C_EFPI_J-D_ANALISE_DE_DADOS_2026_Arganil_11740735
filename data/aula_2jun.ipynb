# Line Plot — gráfico de linhas

```python
df = pd.DataFrame({
    "mes": ["Jan", "Fev", "Mar", "Abr", "Mai", "Jun"] * 2,
    "vendas": [120, 150, 170, 160, 200, 230, 100, 130, 160, 155, 190, 210],
    "produto": ["Produto A"] * 6 + ["Produto B"] * 6
})

"""
df = pd.DataFrame({
    "mes": ["Jan","Jan", "Fev","Fev", "Mar","Mar", "Abr", "Abr","Mai","Mai", "Jun", "Jun"],
    "vendas": [120,100, 150,130, 170,160, 160,155, 200,190, 230, 210],
    "produto": ["Produto A","Produto B","Produto A","Produto B","Produto A","Produto B","Produto A","Produto B","Produto A","Produto B","Produto A","Produto B"]
})
"""


plt.figure(figsize=(9, 5))

sns.lineplot(
    data=df,
    x="mes",
    y="vendas",
    hue="produto",
    marker="^",
    linestyle="-.",
    linewidth=3
)

plt.title("Evolução das vendas por produto", fontsize=16)
plt.xlabel("Mês", fontsize=12)
plt.ylabel("Vendas (€)", fontsize=12)
plt.legend(title="Produto")
plt.tight_layout()

plt.savefig("opt2.png")
plt.show()
```

# Scatter Plot — gráfico de dispersão

```python
df = pd.DataFrame({
    "horas_estudo": [1, 2, 3, 4, 5, 6, 7, 8, 3 , 5, 1],
    "nota": [8, 10, 11, 13, 14, 15, 17, 18, 13, 14, 10],
    "turma": ["A", "A", "A", "B", "B", "B", "A", "B", "C", "C", "C"]
})

plt.figure(figsize=(8, 5))

sns.scatterplot(
    data=df,
    x="horas_estudo",
    y="nota",
    hue="turma",
    style="turma",
    s=150
)

plt.title("Relação entre horas de estudo e nota", fontsize=16)
plt.xlabel("Horas de estudo", fontsize=12)
plt.ylabel("Nota final", fontsize=12)
plt.legend(title="Turma")
plt.tight_layout()
plt.show()
```

# Bar Plot — barras com cores e texturas

```python
df = pd.DataFrame({
    "curso": ["Python", "Java", "C#", "JavaScript"],
    "media": [16, 14, 13, 15]
})

plt.figure(figsize=(8, 5))


ax = sns.barplot(
    data=df,
    x="curso",
    y="media",
    hue="curso",
    palette="Set2"
)


hatches = ["//", "\\\\", "xx", ".."]

"""
for bar, hatch in zip(ax.patches, hatches):
    bar.set_hatch(hatch)
    bar.set_edgecolor("black")
"""




for bar in ax.patches:
    altura = bar.get_height()
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        altura - 1.5,
        f"{altura:.1f}",
        ha="center"
    )




plt.title("Média por curso", fontsize=16)
plt.xlabel("Curso", fontsize=12)
plt.ylabel("Média", fontsize=12)
plt.ylim(0, 20)
plt.tight_layout()
plt.show()
```

# Count Plot — contagem de categorias

```python
df = pd.DataFrame({
    "resultado": ["Aprovado", "Reprovado", "Aprovado", "Aprovado", "Reprovado", "Aprovado"]
})

plt.figure(figsize=(7, 5))

ax = sns.countplot(
    data=df,
    x="resultado",
    palette="pastel",
    hue="resultado",
)

for bar in ax.patches:
    valor = int(bar.get_height())
    ax.text(
        bar.get_x() + bar.get_width() / 2,
        valor + 0.05,
        str(valor),
        ha="center",

    )

plt.title("Número de alunos aprovados e reprovados", fontsize=16)
plt.xlabel("Resultado", fontsize=12)
plt.ylabel("Quantidade", fontsize=12)
plt.tight_layout()
plt.show()
```

# Box Plot

```python
df = pd.DataFrame({
    "turma": ["A"] * 10 + ["B"] * 10,
    "nota": [12, 13, 14, 15, 16, 17, 18, 14, 15, 16,
             8, 10, 11, 13, 14, 15, 15, 16, 17, 19]
})

plt.figure(figsize=(8, 5))

sns.boxplot(
    data=df,
    x="turma",
    y="nota",
    palette="Set3",
    hue="turma",
)

plt.title("Distribuição das notas por turma", fontsize=16)
plt.xlabel("Turma", fontsize=12)
plt.ylabel("Nota", fontsize=12)
plt.tight_layout()
plt.show()
```

# Violin Plot — distribuição detalhada

# Regplot — dispersão com linha de tendência

# Heatmap — quadrados todos do mesmo tamanho

```python
dados = pd.DataFrame(
    [
        [10, 20, 30],
        [40, 50, 60],
        [70, 80, 90]
    ],
    index=["Linha 1", "Linha 2", "Linha 3"],
    columns=["A", "B", "C"]
)

plt.figure(figsize=(6, 6))

sns.heatmap(
    dados,
    annot=True, # colocar o valor no meio
    fmt=".0f",
    cmap="viridis",
    linewidths=1,
    linecolor="white",
    square=True,
    cbar_kws={"label": "Valor"}
)

plt.title("Heatmap com quadrados do mesmo tamanho", fontsize=16)
plt.xlabel("Colunas", fontsize=12)
plt.ylabel("Linhas", fontsize=12)
plt.tight_layout()
plt.show()
```

# "Heatmap" — células com tamanhos diferentes

```python


valores = np.array([
    [10, 20, 30],
    [40, 50, 60],
    [70, 80, 90]
])

# Larguras
x = [0, 1, 1, 6]

# Alturas
y = [0, 1, 2.5, 5]

plt.figure(figsize=(8, 6))

mesh = plt.pcolormesh(
    x,
    y,
    valores,
    cmap="viridis",
    edgecolors="white",
    linewidth=2
)

plt.colorbar(mesh, label="Valor")


for i in range(valores.shape[0]):
    for j in range(valores.shape[1]):
        centro_x = (x[j] + x[j + 1]) / 2
        centro_y = (y[i] + y[i + 1]) / 2

        plt.text(
            centro_x,
            centro_y,
            valores[i, j],
            ha="center",
            va="center",
            color="white",
            fontsize=12,
            fontweight="bold"
        )

plt.title("Heatmap com células de tamanhos diferentes", fontsize=16)
plt.xlabel("Colunas com larguras diferentes", fontsize=12)
plt.ylabel("Linhas com alturas diferentes", fontsize=12)

plt.xticks(
    [(x[i] + x[i + 1]) / 2 for i in range(3)],
    ["A", "B", "C"]
)

plt.yticks(
    [(y[i] + y[i + 1]) / 2 for i in range(3)],
    ["Linha 1", "Linha 2", "Linha 3"]
)

plt.tight_layout()
plt.show()
```

# Heatmap de correlação

```python
iris = sns.load_dataset("iris")

corr = iris.corr(numeric_only=True)

plt.figure(figsize=(8, 6))

sns.heatmap(
    corr,
    annot=True,
    fmt=".2f",
    cmap="coolwarm",
    linewidths=1,
    linecolor="black",
    square=True,
    cbar_kws={"label": "Correlação"}
)

plt.title("Mapa de correlação das variáveis", fontsize=16)
plt.xlabel("Variáveis", fontsize=12)
plt.ylabel("Variáveis", fontsize=12)
plt.tight_layout()
plt.show()
```

# Pairplot — relação entre várias variáveis

```python
iris = sns.load_dataset("iris")

sns.pairplot(
    iris,
    hue="species",
    palette="Set1",
    diag_kind="hist"
)

plt.suptitle("Relações entre variáveis do dataset Iris", y=1.02, fontsize=16)
plt.show()
```

# Heatmap -

```python
import squarify


categorias = ["Python", "Java", "C#", "JavaScript", "PHP"]
valores =   [40,           25,    15,    12,          8  ]

cores = sns.color_palette("Set1", len(valores))

labels = [
    f"{cat}\n{valor}%"
    for cat, valor in zip(categorias, valores)
]

plt.figure(figsize=(10, 6))


squarify.plot(
    sizes=valores,
    label=labels,
    color=cores,
    alpha=1,
    text_kwargs={
        "fontsize": 18,
        "fontweight": "bold",
        "fontstyle": "italic",
        "rotation":45
    }
)

plt.title("Distribuição de utilização por linguagem", fontsize=16)
plt.axis("off")
plt.show()
```
